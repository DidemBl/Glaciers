```python
#Read in the shapefiles from the local directory within a for loop
import os
os.environ['SHAPE_RESTORE_SHX'] = 'YES'
data_files = Path("../Resources/Shapefiles")
shapefile_dfs = []
for data_file in data_files.glob("*.shp"):
    df = gpd.read_file(data_file)
    shapefile_dfs.append(df)
concat_shapefile_df = pd.concat(shapefile_dfs, ignore_index=True)

concat_shapefile_df.head(1)
```

```python
#Determine the total number of rows in the concatenated table
concat_shapefile_df.shape
```

DATA EXPLORATION

```python
#Find the columns of the shapefile
concat_shapefile_df.columns
```

```python
#Copy the shapefile dataframe
shapefile_df = concat_shapefile_df.copy()
shapefile_df.head()
```

```python
# Drop the irrelevant columns
shapefile_df_reduced = shapefile_df[["RGIId","CenLon", "CenLat", "Zmin", "Zmax", "Slope", "Lmax", "Form", "Name", "GN_area", "GN_comp_ye", "geometry"]]
shapefile_df_reduced.head()

```

```python
#Compute the change in area between 1986 and 2020 for each glacier

area_1986_df = shapefile_df_reduced.loc[shapefile_df_reduced["GN_comp_ye"]== "1986", ["RGIId","GN_area","geometry"]]
area_2020_df = shapefile_df_reduced.loc[shapefile_df_reduced["GN_comp_ye"]== "2020", ["RGIId","GN_area","geometry"]] 

joined_area_df = area_1986_df.merge(area_2020_df, on=['RGIId'], how="left")


renamed_df = joined_area_df.rename(columns={'GN_area_x': "1986_area", "GN_area_y": "2020_area", "geometry_x":"polygons_2000", "geometry_y": "polygons_2020"})

renamed_df

area_change = renamed_df["2020_area"] - renamed_df["1986_area"]

renamed_df["change_in_area"] = area_change

renamed_df

```

```python
#Compute the total change in area between 1986 and 2020
total_change = renamed_df["change_in_area"].sum()
total_change
```

```python
#Determine the number of glaciers that shrank in size between 1986 and 2020
glacier_retreat_no = renamed_df.iloc[:,5][renamed_df.iloc[:,5]<0].count()
glacier_retreat_no



```

```python
#Determine the number of glaciers that grew in size between 1986 and 2020
glacier_enlarged = renamed_df.iloc[:,5][renamed_df.iloc[:,5] > 0].count()
glacier_enlarged
```

```python
total_glacier_change_count = renamed_df.iloc[:,5].count()
total_glacier_change_count
```
