```python
#read in data
raw_df = pd.read_csv(Path("../Data/AK_1985_to_2000_overall_glacier_covered_area.csv"))
raw_df.head()
```

```python
#glaciers only
cleaning=raw_df[raw_df['Form'] == 0]
#clarifying column names and converting to Imperial 
cleaning['Id'] = cleaning['RGIId']
cleaning['Height (m)'] = cleaning['Zmax'] - cleaning['Zmin']
cleaning['Height (ft)'] = cleaning['Height (m)'] * 3.281
cleaning['Length (ft)'] = cleaning['Lmax'] * 3.281
cleaning['Area (mi^2)'] = cleaning['GN_area'] / 2.59
cleaning['Year'] = cleaning['GN_comp_ye']
#removing excess columns and sorting by Id
cleaning=cleaning.sort_values(by=["Id", "Year"]).reset_index()
cleaning=cleaning[['Id','Name','Year', 'CenLon', 'CenLat', 'Area (mi^2)', 'Height (ft)', 'Length (ft)']]
cleaning.head(25)
```

```python
#a second csv with data over time for the ai to evaluate. thanks to chatgpt for some of the syntax.
def calculate_changes(df):
    changes = []
    for id_value, group in df.groupby('Id'):
        group_sorted = group.sort_values(by='Year')
        row = {'Id': id_value}
        for i in range(1, len(group_sorted)):
            prev_year = group_sorted.iloc[i-1]['Year']
            curr_year = group_sorted.iloc[i]['Year']
            
            area_change = group_sorted.iloc[i]['Area (mi^2)'] - group_sorted.iloc[i-1]['Area (mi^2)']
            long_change = group_sorted.iloc[i]['CenLon'] - group_sorted.iloc[i-1]['CenLon']
            lat_change = group_sorted.iloc[i]['CenLat'] - group_sorted.iloc[i-1]['CenLat']
            height_change = group_sorted.iloc[i]['Height (ft)'] - group_sorted.iloc[i-1]['Height (ft)']
            length_change = group_sorted.iloc[i]['Length (ft)'] - group_sorted.iloc[i-1]['Length (ft)']
            
            change_column_name = f"change between {prev_year} and {curr_year}"
            row[f"Longitude {change_column_name}"] = long_change
            row[f"Latitude {change_column_name}"] = lat_change
            row[f"Area (mi^2) {change_column_name}"] = area_change
            row[f"Height (ft) {change_column_name}"] = height_change
            row[f"Length (ft){change_column_name}"] = length_change
        changes.append(row)
    df_with_changes = pd.DataFrame(changes)
    return df_with_changes
over_time = calculate_changes(cleaning)
over_time.head(25)
```

```python
#in spot checking, noticed nothing but area changed. decided to clean further.
area_only=over_time
area_only = area_only.loc[:, ~area_only.columns.str.startswith('Longitude')]
area_only = area_only.loc[:, ~area_only.columns.str.startswith('Latitude')]
area_only = area_only.loc[:, ~area_only.columns.str.startswith('Height (ft)')]
area_only = area_only.loc[:, ~area_only.columns.str.startswith('Length (ft)')]
area_only.head(25)
```

```python
#after reviewing this file, we noticed that the glaciers waxed and waned, which was weird. 
#We spoke to Cait, who recommended a UML
#We decided to pull other data into that UML, which has its own script
#
#after that UML, we decided to revisit the SML below
#but by only tracking change between the first year of the data (2000) and last (2020) thus avoiding the waxing and waning
```

```python
#doing just difference between 1986-2020
twodecade=cleaning[cleaning['Year'].isin([1986, 2020])]
td_1986 = twodecade[twodecade['Year'] == 1986]
td_2020 = twodecade[twodecade['Year'] == 2020]
twodecade = pd.concat([td_1986, td_2020])
twodecade=twodecade.sort_values(by=["Id", "Year"]).reset_index(drop=True)
twodecade.head()
```

```python
#calculating changes between 1985 & 2020
td_1986 = twodecade[twodecade['Year'] == 1986]
td_2020 = twodecade[twodecade['Year'] == 2020]
SMLprep = pd.merge(td_1986, td_2020, on=['Id','Name'], suffixes=('_1986', '_2020'))
SMLprep['Area Shrinkage (mi^2)'] = SMLprep['Area (mi^2)_1986']-SMLprep['Area (mi^2)_2020']
SMLprep['Shrinkage'] = SMLprep['Area Shrinkage (mi^2)'] > 0
#since lat, lon, height, and length didnt change as shown earlier, differences not needed
SMLprep['Lat']=SMLprep['CenLat_1986']
SMLprep['Lon']=SMLprep['CenLon_1986']
SMLprep['Height (ft)']=SMLprep['Height (ft)_1986']
SMLprep['Length (ft)']=SMLprep['Length (ft)_1986']
SMLprep=SMLprep[['Id','Name','Lat','Lon','Area (mi^2)_1986','Area (mi^2)_2020','Area Shrinkage (mi^2)','Shrinkage','Height (ft)','Length (ft)']]
SMLprep.head()
```

```python
#print a nice summary
shrink=SMLprep['Shrinkage'].sum()
noshrink=len(SMLprep) - shrink
percentage = round((shrink / len(SMLprep)) * 100, 3)
print(f'Total Glaciers that Shrunk between 1985 and 2020: {shrink}.')
print(f'Total Glaciers that Didnt Shrink between 1985 and 2020: {noshrink}.')
print(f'Total Percentage of Glaciers that Shrunk between 1985 and 2020: {percentage}%.')
```
